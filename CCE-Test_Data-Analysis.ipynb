# CCE test problem
A bottomhole fluid sample from a black oil reservoir at depth of 6126 ft, initial resrvoir pressure of 2338 psia, reservoir temperature of 194 degree farenheit and separator conditions of 100 degree farenheit and 500 psia.
a) Plot oil formation volume factor(Bo) vs pressure to find bubble point pressure
b) Construct density,solution gas-oil ratio and oil viscosity curves versus pressure 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv(r'E:\Dhrumil\Datasets\Separator_test.csv')
df.head()
```

```python
df.shape
```

```python
df
```

```python
# plotting oil FvF vs pressure 
plt.style.use('default')
plt.figure(figsize=(8,4))
plt.plot(df['Pressure'],df['Oil FVF'],label="Bo vs p",color="blue",linewidth=3)
plt.title('CCE test analysis')
plt.xlabel('pressure')
plt.ylabel('Bo')
plt.legend()
plt.grid()
```

```python
# from the graph it can be interpreted that the bubble point of the black oil reservoir is 1500 psi 
# since if we decrease pressure below initial reservoir FVF should increase till Pres_initial=Pb(bubble point)
# Since here the demarcation is observed at 1500 psi, hence bubble point pressure=1500 psi
```

```python
plt.style.use('default')
plt.figure(figsize=(8,4))
plt.plot(df['Pressure'],df['Oil density'],label="density vs p",color="red",linewidth=3)
plt.title('CCE test analysis')
plt.xlabel('pressure')
plt.ylabel('Oil-Density')
plt.legend()
plt.grid()
```

```python
# minimum density observed at 1500 psi pressure i.e 0.589 g/cc
```

```python
plt.style.use('default')
plt.figure(figsize=(8,4))
plt.plot(df['Pressure'],df['Rsi'],label="Rsi vs p",color="green",linewidth=3)
plt.title('CCE test analysis')
plt.xlabel('pressure')
plt.ylabel('Solution gas oil ratio')
plt.legend()
plt.grid()
```

```python
plt.style.use('default')
plt.figure(figsize=(8,4))
plt.plot(df['Pressure'],df['Oil viscosity'],label="Oil-viscosity vs p",color="cyan",linewidth=3)
plt.title('CCE test analysis')
plt.xlabel('pressure')
plt.ylabel('Oil-viscosity')
plt.legend()
plt.grid()
```

```python
# through rsi vs p graph we came to know that rsb=839
df['Gas FVF']=df['Gas FVF'].fillna(0)
df
```

```python
Rsb=839
df['Total FVF']=df['Oil FVF']+(Rsb-df['Rsi'])*df['Gas FVF']
df
```

```python
plt.style.use('default')
plt.figure(figsize=(8,4))
plt.plot(df['Pressure'],df['Total FVF'],label="Bt vs p",color="magenta",linewidth=3)
plt.title('CCE test analysis')
plt.xlabel('pressure')
plt.ylabel('Total formation volume factor')
plt.legend()
plt.grid()
```
