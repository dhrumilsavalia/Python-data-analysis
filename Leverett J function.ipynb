# Leverett J function problem
The properties of an idealized oil reservoir consisting of 4 layers with distinct petrophysical properties. The top of the reservoir is at 8000 ft below the surface and oil water contact is at 8185ft.All the layers have same pore structure but different permeabilities and porosities


# Questions
1) Calculate and plot levertt-J function for the reservoir 
2) Calculate and plot for Layers 2,3 and 4 together with Layer 1


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1=pd.read_excel(r'E:\Dhrumil\Data analysis\Petrophysical_data.xlsx')
df1
```

```python
df2=pd.read_excel(r'E:\Dhrumil\Data analysis\Capillary-Pressure_data.xlsx')
df2
```

```python
density_water=1.036
density_oil=0.822
sigma=35
cos_theta=1
```

```python
k_layer1=df1.loc[2,'Layer1']
h_layer1=df1.loc[1,'Layer1']
k_layer2=df1.loc[2,'Layer2']
h_layer2=df1.loc[1,'Layer2']
k_layer3=df1.loc[2,'Layer3']
h_layer3=df1.loc[1,'Layer3']
k_layer4=df1.loc[2,'Layer4']
h_layer4=df1.loc[1,'Layer4']
```

```python
df2['J_Layer1']=(0.21645*df2['Pc'])*np.sqrt(k_layer1/df1.loc[3,'Layer1'])/sigma


```

```python
df2['Pc_Layer2']=(df2['J_Layer1']*sigma/0.21645)*np.sqrt(df1.loc[3,'Layer2']/k_layer2)
df2['Pc_Layer3']=(df2['J_Layer1']*sigma/0.21645)*np.sqrt(df1.loc[3,'Layer3']/k_layer3)
df2['Pc_Layer4']=(df2['J_Layer1']*sigma/0.21645)*np.sqrt(df1.loc[3,'Layer4']/k_layer4)
df2
```

```python
new=df2.drop(columns=['J_Layer2','J_Layer3','J_Layer4'])
new
```

```python
# leverett J plot for reservoir
plt.style.use('default')
plt.figure(figsize=(8,4))
plt.plot(new['Sw'],new['J_Layer1'],label='J',color='blue',linewidth=3)
plt.xlabel('Sw')
plt.ylabel('J(Sw)')
plt.title('Leverett- J function for reservoir')
plt.legend()
plt.grid()
```

```python
# capillary pressure curves for layer1,2,3 and 4
plt.style.use('default')
plt.figure(figsize=(8,4))
plt.plot(new['Sw'],new['Pc'],label='Layer1',color='blue',linewidth=1)
plt.plot(new['Sw'],new['Pc_Layer2'],label='Layer2',color='red',linewidth=1)
plt.plot(new['Sw'],new['Pc_Layer3'],label='Layer3',color='green',linewidth=1)
plt.plot(new['Sw'],new['Pc_Layer4'],label='Layer4',color='magenta',linewidth=1)
plt.title('Capillary pressure curve for 4 layers')
plt.xlabel('Sw')
plt.ylabel('Pc(Sw)')
plt.legend(loc='best')
plt.grid()
```
